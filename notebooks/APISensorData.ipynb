```python
import urllib.request
import json
import glob
import pandas as pd
import numpy as np
```

```python
# this cell gets data
URL = "http://api.aberdeenspaceport.com/api/v2/data?lat=56.964&lon=-2.212&start_time=2019-05-05T10:00:00&end_time=2019-05-25T11:00:00" 
response = urllib.request.urlopen(URL)
data = json.loads(response.read())
print(data)
```

```python
# this cell gets data
URL = "http://api.aberdeenspaceport.com/api/v1/data?lat=56.964&lon=-2.212&start_date=2019-05-05T10:00:00&end_date=2019-05-25T11:00:00" 
response = urllib.request.urlopen(URL)
data = json.loads(response.read())
print(data)
```

```python
columns = ['index', 'TimeStamp','P1','P2', 'Humidity', 'Pressure', 'Temperature']
df_P = pd.DataFrame()
print(df_P)
i=0
for values in data:
    Sensor_id_val = values.get('11407')
    sensor_vals = values.get('readings')
    Index = sensor_vals[0]['value']
    Time Stamp = sensor_vals[1]['value']
    P1 = sensor_vals[2]['value']
    P2 = sensor_vals[3]['value']
    Humidity = sensor_vals[4]['value']
    Pressure = sensor_vals[5]['value']
    Temperature = sensor_vals[6]['value']
    row = pd.Series([i,sensor_id_vals,Index,TimeStamp,P1, P2, Humidity, Pressure, Temperature])    
    df_row = pd.DataFrame(row).transpose()    
    df_P = pd.concat([df_P,df_row],axis=0)
    i=i+1
  
df_P = pd.DataFrame(data= df_P.values, columns = columns )    
print(df_P)
    
```
